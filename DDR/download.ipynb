```python
import datetime
import json
import pickle

import pandas as pd
```

```python
pathways = pd.read_csv('2021_0211_Pan Cancer DDR gene sets.csv').fillna('')
```

```python
pathways
```

```python
k_pw_v_genes = {}

for c in pathways.columns:
    genes = [x for x in pathways[c].values if len(x)]
    if len(genes):
        k_pw_v_genes[c] = genes
```

```python
now =  datetime.datetime.now()
readme_dict = {
    'date_accessed': now.strftime("%b %d %Y %H:%M:%S"),
    'pathways': len(k_pw_v_genes),
}

results = {
    'access_log': readme_dict,
    'data': k_pw_v_genes
}

with open(f'results.json', 'w') as fp:
        json.dump(results, fp)
        
pickle.dump(results, open('results.pkl', 'wb'))
```
